```python
                        #Assignment2 Week 3
# This program takes User Name and Subject Marks and prints a Marks Sheet

# user input  
Name       =  input("Enter your Name =")
Roll_no    =  input("Enter your Roll_no =")
Exam_Year  =  input("Enter your Exam_Year =")
Semester   =  input("Enter your Semester =")
Report_Date=  input("Enter your Report_Date =")
# here user input only take int  datatype  

OPERATING_SYSTEM = int(input("Enter the Mark in OPERATING SYSTEM (out of 100)= "))
VISUAL_PROGRAMMING= int(input("Enter the Mark in VISUAL PROGRAMMING (out of 100)= "))
COMPILER_CONSTRUCTION= int(input("Enter the Mark in COMPILER CONSTRUCTION (out of 100)= "))
COMPUTER_NETWORKS= int(input("Enter the Mark in COMPUTER NETWORKS (out of 100)= "))
ARTIFICIAL_INTELLIGENCE= int(input("Enter the Mark in ARTIFICIAL INTELLIGENCE (out of 100)= "))

# we sum the obtain mark beacuse sooner we have to find the percetage of sum of marks
Obtained_Mark = OPERATING_SYSTEM+VISUAL_PROGRAMMING+COMPILER_CONSTRUCTION+COMPUTER_NETWORKS+ARTIFICIAL_INTELLIGENCE

# here is tool marks 
Total_Marks = 500

# now we want to find the percentage to of mark .we this in the condtion where grade will be assign 
percent_Marks = (Obtained_Mark / Total_Marks) * 100

# grade
grade = "F"

# condition
if percent_Marks > 90 :
    grade = "A"
elif( percent_Marks < 90 and percent_Marks >80 ):
    grade = "B"
elif( percent_Marks < 80 and percent_Marks >70 ):
    grade = "C"
elif( percent_Marks < 70 and percent_Marks >60 ):
    grade = "D"
else :
    grade = "F"
# now we are going to print all data
print("*******************************************************************")


print("|\tName \t\t\t= ",Name,)
print("|\tRoll_no \t\t= ",Roll_no)
print("|\tExam_Year \t\t= ",Exam_Year)
print("|\tSemester \t\t= ",Semester)
print("|\tReport_Date \t\t= ",Report_Date)
print("|")
print("|\tOPERATING_SYSTEM \t= ",OPERATING_SYSTEM,"/ 100")
print("|\tVISUAL_PROGRAMMING \t= ",VISUAL_PROGRAMMING,"/ 100")
print("|\tCOMPILER_CONSTRUCTION   = ",COMPILER_CONSTRUCTION,"/ 100")
print("|\tCOMPUTER_NETWORKS \t= ",COMPUTER_NETWORKS,"/ 100")
print("|\tARTIFICIAL_INTELLIGENCE = ",ARTIFICIAL_INTELLIGENCE,"/ 100")
print("|")
print("|\tObtained_Mark \t\t= ",Obtained_Mark,"/",Total_Marks)
print("|\tpercent_Marks \t\t= ",percent_Marks)
print("|\tTotal_Marks \t\t= ",Total_Marks)
print("|\tTotal_Marks \t\t= ",grade)






```
